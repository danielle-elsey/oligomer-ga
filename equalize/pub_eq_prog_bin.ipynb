## Methods

```python
# takes dataframe with cols: run, gen, filename, alpha, mu, vol
# returns dataframe with cols: run, gen, filename, alpha_term, mu_term, vol

def convert_to_clausius_terms(df):
    df2 = pd.DataFrame()

    # loop over rows
    for x in range(df.shape[0]):

        # alpha in bohr^3, mu in debye, vol in A^3
        run = df.iloc[x, 0]
        gen = df.iloc[x,1]
        filename = df.iloc[x, 2]
        alpha = df.iloc[x,3]
        mu = df.iloc[x,4]
        vol = df.iloc[x,5]

        # convert polarizability (volume) to SI units (bohr^3 to m^3)    
        si_alpha_prime = alpha * (5.29177E-11)**3

        # convert dipole moment to SI units (debye to C*m)
        # 1 debye = 3.3E-30 C*m
        si_mu = mu * 3.3356E-30

        # convert volume to SI units (A^3 to m^3)
        si_vol = vol * 1E-30

        # define molecular density N/V (assume packing of 68%)
        n_v = 0.68/si_vol    

        # calculate Clausius-Mossotti equation term values (assume temperature of 300K)
        a_term = n_v * (si_alpha_prime * 4 * sc.pi * sc.epsilon_0)/(3 * sc.epsilon_0)
        m_term = n_v * (si_mu**2)/(9 * sc.epsilon_0 * sc.Boltzmann * 300)

        claus_sum = a_term + m_term

        # solve C-M eqn for dielectric constant
#         dielec = (2 * claus_sum + 1)/(1 - claus_sum)

        temp_df = pd.DataFrame({'run':[run], 'gen':[gen], 'filename':[filename], 'alpha_term':[a_term], 'mu_term':[m_term], 'vol':[vol]})
        df2 = df2.append(temp_df, ignore_index=True)

    return df2
```

```python
# takes any dataframe with cols: alpha_term, mu_term
# returns list of 6 dataframes for each of 6 bounded bins (same cols as input df)

# note: bins are organized in two rows L-R, 1-3 and 4-6 (NOT clockwise)
def make_bins_6(df):
    # define bin boundaries
    # horizontal boundary between bins 1/2/3 and 4/5/6
    median = df['alpha_term'].median()

    # two vertical boundaries between bins 1/5 and 2/4; bins 2/4 and 3
    df_hz_max = df['mu_term'].max()
    bound1 = df_hz_max/3
    bound2 = bound1*2
    
    # make bin dataframes    
    bin1 = df[(df['mu_term'] >= 0) & (df['mu_term'] < bound1) & (df['alpha_term'] >= median)]
    bin2 = df[(df['mu_term'] >= bound1) & (df['mu_term'] < bound2) & (df['alpha_term'] >= median)]
    bin3 = df[(df['mu_term'] >= bound2) & (df['alpha_term'] >= median)]

    bin4 = df[(df['mu_term'] >= 0) & (df['mu_term'] < bound1) & (df['alpha_term'] < median)]
    bin5 = df[(df['mu_term'] >= bound1) & (df['mu_term'] < bound2) & (df['alpha_term'] < median)]
    bin6 = df[(df['mu_term'] >= bound2) & (df['alpha_term'] < median)]

    binned_df_list =  [bin1, bin2, bin3, bin4, bin5, bin6]
    return(binned_df_list)
```

```python
# takes any dataframe with cols: gen, alpha_term, mu_term
# returns dataframe with cols: gen, bin_3_count

# note: bin 3 is the upper right most of 6 bins arranged in a 2x3 grid
def get_bin_3_counts(df):
    max_gen = df['gen'].iloc[-1]
    
    df3 = pd.DataFrame()    
    for gen in range(max_gen):    
        # get each generation individually
        df2 = df.copy()
        df2 = df2[df2['gen'] == gen]
        
        # define bin boundaries
        # horizontal boundary between bins 1/2/3 and 4/5/6
        median = df2['alpha_term'].median()
        # two vertical boundaries between bins 1/5 and 2/4; bins 2/4 and 3
        df2_hz_max = df2['mu_term'].max()
        bound1 = df2_hz_max/3
        bound2 = bound1*2
    
        # make bin 3 dataframe    
        bin3 = df2[(df2['mu_term'] >= bound2) & (df2['alpha_term'] >= median)]
        
        # get number of polymers in bin 3
        bin_3_count = len(bin3.index)
        
        temp_df = pd.DataFrame({'gen':[gen], 'bin_3_count':[bin_3_count]})
        df3 = df3.append(temp_df, ignore_index=True)

    return(df3)
```

```python
# takes any dataframe with cols: gen, alpha_term, mu_term
# returns dataframe with same cols, only rows that fell in "bin 3"

# note: bin 3 is the upper right most of 6 bins arranged in a 2x3 grid
def get_bin_3_data(df):
    # define bin boundaries
    # horizontal boundary between bins 1/2/3 and 4/5/6
    median = df['alpha_term'].median()
    # two vertical boundaries between bins 1/5 and 2/4; bins 2/4 and 3
    df_hz_max = df['mu_term'].max()
    bound1 = df_hz_max/3
    bound2 = bound1*2

    # make bin 3 dataframe    
    bin3 = df[(df['mu_term'] >= bound2) & (df['alpha_term'] >= median)]
    bin3 = bin3.reset_index(drop=True)

    return(bin3)
```

```python
# helper function for top_ten
# takes any dataframe with col: filename
# returns new dataframe with additional cols: mono1, mono2, seq

def split_filename(df):
    df2 = df.copy()

    new = df2['filename'].str.split('_', expand=True)
    df2['mono1'] = new[0]
    df2['mono2'] = new[1]
    df2['seq'] = new[2]
    
    return df2
```

```python
# takes any dataframe with cols: mono1, mono2, seq
# returns list of 2 new dataframes with cols: [x_mono, x_mono_count, x_mono_perc] and [x_seq, x_seq_count, x_seq_perc] 
# where x is specified df_name

# removes any 'false' monomer instances in homopolymers (e.g. mono1 with seq '111111')

def get_bin_counts(df, df_name):

    # split filename column
    df_split = split_filename(df)
    
    # replace 'false' monomer instances in homopolymers (e.g. mono1 with seq '111111') with 'x'
    df_split.loc[df_split['seq'].eq('000000'), 'mono1'] = 'x'
    df_split.loc[df_split['seq'].eq('111111'), 'mono2'] = 'x'

    # make series objects of each part of filename with accompanying counts
    # note that both monomer columns are combined before creating series with value counts
    count_mono = df_split[['mono1', 'mono2']].unstack().reset_index(drop=True).value_counts()
    count_seq = df_split['seq'].value_counts()
    
    # make series objects into dataframes and rename columns appropriately
    count_mono = pd.DataFrame(count_mono.reset_index())
    count_mono.columns = ['%s_mono' % df_name, '%s_mono_count' % df_name]

    count_seq = pd.DataFrame(count_seq.reset_index())
    count_seq.columns = ['%s_seq' % df_name, '%s_seq_count' % df_name]
    
    # remove monomer 'x' row from dataframe
    count_mono = count_mono[count_mono['%s_mono' % df_name] != 'x'].reset_index(drop=True)
    
    # make percentage column in each dataframe  
    count_mono['%s_mono_perc' % df_name] = count_mono['%s_mono_count' % df_name]/count_mono['%s_mono_count' % df_name].sum()*100
    count_seq['%s_seq_perc' % df_name] = count_seq['%s_seq_count' % df_name]/count_seq['%s_seq_count' % df_name].sum()*100

    return [count_mono, count_seq]  
```

## Generate Plot

Set global foant sizes for plotting

```python
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=MEDIUM_SIZE)  # fontsize of the figure title
```

Make plot.

```python
fig, axs = plt.subplots(1, 4, sharey=True, dpi=300)

# set figure axes
xlimits = (0, 420)
xticks = [0, 120, 240, 360]

ylimits = (0.28, 0.40)
yticks = [0.30, 0.34, 0.38]


# 100 gen
axs[0].scatter(x=run_terms_cf3[:3200]['mu_term'], y=run_terms_cf3[:3200]['alpha_term'], s=8, c='#4394f0', edgecolor='none', alpha = 0.07)
axs[0].set_title('100 Generations')
axs[0].set(ylim=ylimits, xlim=xlimits)
axs[0].set_xlabel('')
axs[0].set_ylabel('')
axs[0].set_xticks(xticks)
axs[0].set_yticks(yticks)

# 200 gen
axs[1].scatter(x=run_terms_cf3[:6400]['mu_term'], y=run_terms_cf3[:6400]['alpha_term'], s=8, c='#4394f0', edgecolor='none', alpha = 0.07)
axs[1].set_title('200 Generations')
axs[1].set(ylim=ylimits, xlim=xlimits)
axs[1].set_xlabel('')
axs[1].set_xticks(xticks)

for tic in axs[1].yaxis.get_major_ticks():
    tic.tick1On = tic.tick2On = False

# 300 gen
axs[2].scatter(x=run_terms_cf3[:9600]['mu_term'], y=run_terms_cf3[:9600]['alpha_term'], s=8, c='#4394f0', edgecolor='none', alpha = 0.07)
axs[2].set_title('300 Generations')
axs[2].set(ylim=ylimits, xlim=xlimits)
axs[2].set_xlabel('')
axs[2].set_xticks(xticks)

for tic in axs[2].yaxis.get_major_ticks():
    tic.tick1On = tic.tick2On = False

# 400 gen
axs[3].scatter(x=run_terms_cf3['mu_term'], y=run_terms_cf3['alpha_term'], s=8, c='#4394f0', edgecolor='none', alpha = 0.07)
axs[3].set_title('400 Generations')
axs[3].set(ylim=ylimits, xlim=xlimits)
axs[3].set_xlabel('')
axs[3].set_xticks(xticks)

for tic in axs[3].yaxis.get_major_ticks():
    tic.tick1On = tic.tick2On = False

# define and mark bin boundaries
gen_count = 3200
for ax in axs:
    # vertical bin boundaries at equal thirds between 0 and max dipole value
    df_hz_max = run_terms_cf3[:gen_count]['mu_term'].max()
    bound1 = df_hz_max/3
    bound2 = bound1*2
    # horizontal bin boundary at median polarizability value
    median = run_terms_cf3[:gen_count]['alpha_term'].median()
    
    # draw lines
    ax.hlines(y=median, xmin=0, xmax=df_hz_max, linewidth=0.5)
    ax.axvline(x=bound1, color='black', linewidth=0.5)
    ax.axvline(x=bound2, color = 'black', linewidth=0.5)
    ax.axvline(x=df_hz_max+1, color = 'black', linewidth=0.5)

    # label bin numbers
    cent_col_1 = bound1/2
    cent_col_2 = (bound2-bound1)/2 + bound1
    cent_col_3 = (df_hz_max-bound2)/2 + bound2
    ax.text(cent_col_1, 0.385, 'C', fontsize = 6)
    ax.text(cent_col_2, 0.385, 'B', fontsize = 6)
    ax.text(cent_col_3, 0.385, 'A', fontsize = 6)
    ax.text(cent_col_1, 0.286, 'F', fontsize = 6)
    ax.text(cent_col_2, 0.286, 'E', fontsize = 6)
    ax.text(cent_col_3, 0.286, 'D', fontsize = 6)
    
    # increase by 100 generations (32 polymers/gen * 100 gen)
    gen_count = gen_count + 3200

top = 0.3
bottom = 0.12

left = 0.11
right = 0.95
    
# label axes
fig.text((left+right)/2, 0.02,'Dipole Moment Term', ha='center')
fig.text(0.01, (top+bottom)/2,'Polarizability Term', va='center', rotation='vertical')

plt.subplots_adjust(top=top, bottom=bottom, left=left, right=right, wspace=0.07)

# plt.savefig('equalize_cf3_progressive_bins.pdf', transparent=False, bbox_inches='tight')
# plt.savefig('equalize_cf3_progressive_bins.png', transparent=False, bbox_inches='tight')

plt.show()
```
